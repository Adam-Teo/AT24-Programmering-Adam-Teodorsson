## 2. Dice simulation (*)

Simulate 1000000 dice rolls and save the number of ones, twos, ..., sixes in a dictionary. Then print them out in the terminal. (*)

<details>

<summary>Answer</summary>

```
1: 166481
2: 166844
3: 166154
4: 167400
5: 167260
6: 165861
```

</details>


````python
#09.02
from random import randint as rng 

dice = { n : 0 for n in range(1, 7) }

for n in range(10**6):
    dice[rng(1, 6)] += 1

for key, val in dice.items():
    print(f"{key}: {val}  ")
````

## 4. Morse code (**)

Read in the file morse.txt, save it in a dictionary and create a function that lets the user input a message to get it translated to morse code. For example  (**)

```python

print(morse("SOS"))
print(morse("POKEMON"))

...---...
.------.-.------.
```

````python
#09.04
cipher = dict()
path = "C:/Code/python-programming-Adam-Teodorsson/" 
with open(path+"Data/morse.txt", "r") as file:
    for e in file.readlines():
        cipher[e[0]] = e[3:-1]

def morse(txt:str)->str: 
    return "".join([cipher[c] for c in txt.upper()])

print(morse("SOS"))
print(morse("POKEMON"))
print(morse("ZeRoX"))
````
